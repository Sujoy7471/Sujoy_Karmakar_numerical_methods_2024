### * 2.1:
#### We know for fixed point iteration, the modified function g(x) when expanded about the root of original function f(x), i.e, about x*, we get the series,

$g(x^*) = g(x^{k-1}) + (x^* - x^{k-1})g'(x^{k-1}) + \frac{(x^* - x^{k-1})^2}{2}g''(x^{k-1}) + .... $

$or, g(x^{k-1}) - g(x^*) = - (x^* - x^{k-1})g'(x^{k-1}) - \frac{(x^* - x^{k-1})^2}{2}g''(x^{k-1}) + .... $

Now, as for fixed point iteration, $ x^k = g(x^{k-1})$, putting in the above expression, 

$x^{k} - x^* = (x^{k-1} - x^*)g'(x^{k-1}) - \frac{(x^* - x^{k-1})^2}{2}g''(x^{k-1}) + .... $

From the above expression, we can see that though$g'(x^{k-1}) = 0$, it does not guarantee that we will get exact solution as the higher order terms are also present there. But it will speed up the process.

### * 2.2:
Fron the above expression, we get (neglecting higher order terms)

$x^{k} - x^* = - \frac{(x^* - x^{k-1})^2}{2}g''(x^{k-1}) $

$or, |x^{k} - x^*| = |{(x^* - x^{k-1})^2}||\frac{g''(x^{k-1})}{2}| $

So, the above expression shows that the convergence will be quadrantic (proved).

```python
import matplotlib.pyplot as plt
import numpy as np

err_dfx = np.zeros(16)                                    # Define the array
h = np.zeros(16)

for i in range(16):
    x = 2.5             # Value of x
    h[i] = 10**(-i)     # Step size
    exact_derivative = 4 * x**3 - 30 * x**2 + 70 * x - 50  # Exact derivative at x
    err_dfx[i] = abs(rich_c(x, h[i]) - exact_derivative)   # Absolute error

#Plotting the figure
plt.figure()
plt.plot(h, err_dfx, ls='-', marker="o", label="Richardson Central")
plt.xlabel("Step Size (h) (in log scale)", fontsize=20)
plt.ylabel("Error (in log scale)", fontsize=20)
plt.xscale('log')
plt.yscale('log')
plt.title("Plot to Get Lowest Error")
plt.legend()
plt.grid()
plt.show()
```

```python
def bisection(a,b,fn,tol,n):                              ##Define the function
    """
    a = lower limit
    b = upper limit
    fn = function
    tol = tolerence
    n = max number of iterations
    """
    ##Iteration recorder
    j = 0

    if fn(a)*fn(b) >=0:
        print("Bad choice of a, b!!!!!!!!!")
        return
    else:
        for i in range(n):
            ##Defining iteration variables (function)
            fl, fu = fn(a), fn(b)
            z = (a+b)/2
            if fn(z) == 0 or abs(fl - fu) < tol:         ##Check if tolerance is met or exact root found
                break
            if fl*fn(z) <0:                              ##Updates upper lim
                b = z 
            if fu*fn(z) <0:                              ##Update lower lim
                a = z
            j = j+1
        return z,j
    


##Finding the roots
root, iteration = bisection(0.6,1.2,fx,10**-8, 200)
print(f"The root is: {root:0.8f},   \tThe iteration number is: {iteration}")
root, iteration = bisection(1.6,2.3,fx,10**-8, 200)
print(f"The root is: {root:0.8f},   \tThe iteration number is: {iteration}")
root, iteration = bisection(2.6,3.2,fx,10**-8, 200)
print(f"The root is: {root:0.8f},   \tThe iteration number is: {iteration}")
root, iteration = bisection(3.5,4.2,fx,10**-8, 200)
print(f"The root is: {root:0.8f},   \tThe iteration number is: {iteration}")

```
